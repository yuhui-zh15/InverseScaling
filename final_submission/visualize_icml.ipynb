```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
from colors import red, green, blue

# Change font to Gill Sans
font_path = '/System/Library/Fonts/Supplemental/GillSans.ttc'
prop = fm.FontProperties(fname=font_path)

# Change the font for all text elements
plt.rcParams['font.family'] = prop.get_name()


def plot(title, X, Ys, legends, shapes=None, offsets=None, legend_loc='upper left'):
    plt.figure(figsize=(10, 6.6), dpi=300)
    # plt.figure(figsize=(10, 5.4), dpi=300)
    markers = ['o', 's', '^', 'h'] 
    colors = ["orange", "red", "blue", "darkgreen"]
    # markers = ['o', 'o', 'o', '^', '^', '^'] 
    # i want a continuous color gradient
    # colors = ["#ff0000", "#ff8000", "#ffcc00", "#00dd00", "#0000ff", "#8000ff"]
    if shapes is None:
        shapes = [None] * len(Ys)
    if offsets is None:
        offsets = [0] * len(Ys)
    for i, (Y, legend, shape, offset) in enumerate(zip(Ys, legends, shapes, offsets)):
        plt.plot(X[:len(Y)], Y, linewidth=2.5, marker=markers[i], markersize=10, label=legend, color=colors[i])
        # put a text near the last point
        if shape is not None:
            plt.text(len(X) - 1 - 0.25, Y[-1] + 2 + offset, shape, fontsize=20, color='black', ha='center')

    # Use a white background
    plt.grid(color='white', linestyle='--')
    plt.gca().set_facecolor('white')

    # Add labels and title to the plot
    plt.xlabel("Scale", fontsize=25)
    plt.ylabel("Accuracy (%)", fontsize=25)
    plt.title(title, fontsize=30)

    # Set the Y-axis range to 0-1, with 0.1 ticks, add a small line for each tick
    plt.ylim(0, 100)
    # plt.ylim(20, 100)
    # plt.yticks(np.arange(20, 120, 20))
    plt.grid(axis='y', color='lightgray', linestyle='--')

    # Add a horizontal line y=0.5
    plt.axhline(y=50, color='gray', linestyle='--', label = "Random")

    # Change fontsize of x tick label
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)

    # Put the legend in left upper corner, set transparent 50%
    plt.legend(loc=legend_loc, fontsize=20, framealpha=0.5)
    # ) # 11) # 20

    # Show the plot
    plt.show()
```

```python
def task1_score(x): # Normal QA
    return 0.025 * x + 0.75

def task2_score(x): # Negation understanding
    return 1 / (1 + np.exp(-x))

def task2_accuracy(x): # Negation understanding
    return (1 / (1 + np.exp(-x))) * 0.5 + 0.5

def composed_score(task1, task2): # Negated QA
    return task1 * task2 + (1-task1) * (1-task2)

plt.figure(figsize=(6, 3), dpi=300)

x = np.linspace(-10,10,100)
plt.plot(x, task1_score(x), label="Task 1: Question Answering", linestyle="dashed")
plt.plot(x, task2_accuracy(x), label="Task 2: Negation Understanding",  linestyle="dashed")
plt.plot(x, composed_score(task1_score(x), task2_score(x)), label="Composed Task: NeQA")
plt.legend()

plt.xlabel("Scale")
plt.ylabel("Accuracy")

# hide the x axis number
plt.xticks([])

# set y from 0 to 1
plt.ylim(-0.05, 1.05)

plt.title("Simulation of Task Decomposition")
```
